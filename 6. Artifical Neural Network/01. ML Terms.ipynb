## **01. 머신러닝 용어 이해하기**
---

### **01. 머신러닝 모델의 평가**
* 모델을 평가하기 위해서 훈련용, 검증용, 테스트용 이렇게 세가지로 분리함. 
    * 각각 Training, Validation, Testing
* 검증용 데이터(Validation)는 모델의 성능을 평가하는 것이 아니라, 모델의 성능을 조정하기 위한 용도임. 주로 **하이퍼파라미터**를 조정에 사용됨. (tuning)
    * 하이퍼파라미터 : 모델의 성능에 영향을 주는 매개변수
    * 파라미터: 가중치, 편향과 같은 학습을 통해 바뀌어져가는 변수 
    * 하이퍼파라미터는 사용자가 직접 정해주는 변수, 파라미터는 모델이 학습하는 과정에서 얻어지는 값. 
* 훈련용 데이터로 훈련을 모두 시킨 모델의 정확도 검증에 검증용 데이터 사용 
* 검증데이터와 테스트 데이터를 나눌만큼 데이터가 충분하지 않다면 k-fold 교차 검증을 통해 실시할 수 있음. 

### **02. 분류(Classification)와 회귀(Regression)**
* 회귀: 선형회귀 // 분류 : 로지스틱 회귀
* 분류의 경우 **이진 분류(Binary Classification)**, **다중 클래스 분류(Multi-Class Classification)**으로 나뉠 수 있음. 

#### **1) 이진 분류 문제(Binary Classification)**
* 주어진 입력에 대해서 둘 중 하나의 답을 정하는 문제
* ex) 시험 성적에 대해서 합격, 불합격인지 판단하고 메일이 정상, 스팸인지 판단 

#### **2) 다중 클래스 분류(Multi-Class Classification)**
* 주어진 입력에 대해서 세 개 이상의 정해진 선택지 중에서 답을 정하는 문제 
* ex) 과학, 영어, IT, 만화와 같이 4개의 레이블이 있음. 이때 4개의 선택지를 카테고리 또는 클래스라고 하며, 주어진 입력으로부터 정해진 클래스 중 하나로 판단하는 문제

#### **3) 회귀 문제**
* 분류 문제처럼 0 또는 1이나 과학, IT등과 같이 분리된 (비연속적인) 답이 결과가 아니라 **연속된** 값을 결과로 가짐. 
* ex) 시계열 데이터를 이용한 주가예측, 생산량 예측, 지수 예측등이 속함. 

### **03. 지도 학습(Supervised Learning)과 비지도 학습(Unsupervised Learning)**
#### **1) 지도 학습**
* 레이블(Label)이라는 정답과 함께 학습하는 것을 말함. 
* $y$, 실제값 등이 레이블로 존재하면, 예측값과 실제값의 차이인 오차를 줄이는 방식으로 학습하게 되는데 예측값은 $\widehat{y}$ 과 같이 표현함. 
#### **2) 비지도 학습**
* 기본적으로 목적 데이터(또는 레이블)이 없는 학습 방법
* 군집(clustering)이나 차원 축소와 같은 학습 방법이 있음. .
#### **3) 강화 학습**
* 어떤 환경 내에서 정의된 에이전트가 현재 상태를 의식하여, 선택 가능한 행동들 중 보상을 최대화하는 행동 혹은 순서를 선택함.

### **05. 혼동 행렬(Confusion Matrix)**
* 정확도: 맞춘 문제수를 전체 문제수로 나눈 값
* 양성, 음성을 구분하는 이진 분류가 있다고 했을때, 혼동 행렬은 다음과 같음.


|-|참|거짓|
|------|---|---|
|참|TP|FN|
|거짓|FP|TN|

* 여기서 TP(True Positive), TN(True Negative), FP(False Positive), FN(False Negative)라고 함. 
* True는 정답을 맞춘 경우, False는 정답을 맞추지 못한 경우
* Positive는 각각 제시했던 정답임. TP는 양성이라고 대답했을때, 실제로 양성이라서 정답을 맞춘 경우, TN은 음성이라고 대답했을때 실제로 음성이라서 정답을 맞춘 경우
* FP는 양성이라고 대답했는데, 음성이라서 틀린 경우 , FN은 음성이라고 대답했는데 양성이라서 정답을 틀린 경우

#### **1) 정밀도(Precision)**
* 양성이라고 대답한 전체 케이스에 대한 TP의 비율
* $$ 정밀도 =  \frac{TP}{TP+FP} $$

#### **2) 재현률(Recall)**
* 실제값이 양성인 데이터의 전체 개수에 대해서 TP의 비율
* 양성인 데이터 중에 얼마나 양성인지를 예측했는지 나타냄
* $$ 재현률 =  \frac{TP}{TP+TN} $$

### **06. 과적합(Overfitting)과 과소적합(Underfitting)
#### **(1) 과적합**
* 훈련 데이터를 과하게 학습한 경우를 말함. 
* 훈련데이터에 대해서 과하게 학습하면 테스트 데이터나 실제 서비스에서 데이터에 대해 정확도가 좋지 않은 현상 
* 과적합 상황에서는 훈련 데이터에 대해서 오차가 낮지만, 테스트 데이터에 대해서 오차가 높아지는 현상 
* 과적합을 방지하기 위해서 ealry stop, 즉 오차가 증가하지 전이나 정확도가 감소하기 전에 훈련을 멈추는 것임. 

#### **(2) 과소적합**
* 테스트 데이터의 성능이 올라갈 여지가 있음에도 훈련을 덜 한 상태를 반대로 과소적합이라고 함. 
* 과소적합은 훈련 자체가 부족한 상태이므로 과대 적합과 달리 훈련 데이터에 대해서도 보통 정확도가 낮음


* 딥러닝을 할때 과적합을 막을 수 있는 드롭아웃(Dropout), 조기 종료(Early Stopping)과 같은 방법이 존재함.
